```python
The Exception class is used as a base class when creating a custom exception 
in order to inherit its properties and methods. 
When a new custom exception class is created, it can inherit the properties and methods of the Exception class, 
which can make it easier to handle and manage exceptions in a program.
Using the Exception class as a base class can help to maintain consistency with other exceptions in the program. 
Since all built-in exceptions and most custom exceptions inherit from the Exception class, 
using it as a base class for a new custom exception can help to ensure that it behaves in a 
consistent way with other exceptions in the program.
```

```python
#ZeroDivisionError
import sys
def zde():
    try:
        a = 7/0
        print(a)
    except ArithmeticError as e:
        print(e)
zde()
```

```python
LookupError Exception is the Base class for errors raised when something can't be found. 
The base class for the exceptions that are raised when a key or 
index used on a mapping or sequence is invalid: IndexError, KeyError.

Examples are below:
```

```python
#index_error
l=[1,2,3,"xyz"]
for i in range(5):
    print(l[i])
```

```python
#key_error
d1 = {1:'abc',
      'd':234,
      5:'efg'
}

print(d1[3])
```
